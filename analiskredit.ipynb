```python
def klasifikasi_plafon(x) :
    if 60000000 <= x < 70000000 :
        x = 'kurang dari 70 juta'
    elif 50000000 <= x < 60000000 :
        x = 'kurang dari 60 juta' 
    elif 40000000 <= x < 50000000 :
        x = 'kurang dari 50 juta' 
    elif 30000000 <= x < 40000000 :
        x = 'kurang dari 40 juta' 
    elif 20000000 <= x < 30000000 :
        x = 'kurang dari 30 juta' 
    elif 10000000 <= x < 20000000 :
        x = 'kurang dari 20 juta' 
    elif 5000000 <= x < 10000000 :
        x = 'kurang dari 10 juta' 
    elif 1000000 <= x < 5000000 :
        x = 'kurang dari  5 juta' 
    elif 0 < x < 10000000 :
        x = 'kurang dari  1 juta' 
    else :
        x = 'tidak mendapat plafon'
    return x
```

```python
z = df['plafon'].map(klasifikasi_plafon)
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(15,8))


y = df['status_angsuran'].value_counts().index
y_pos = np.arange(len(y))
x = df['status_angsuran'].value_counts().values
tot = df['status_angsuran'].value_counts().values.sum()

ax = sns.countplot(x="status_angsuran", data=df)
ax.set_xticklabels(y,fontsize=14)
ax.set_xlabel('')
ax.set_title("Status Angsuran",fontsize=18)

for a in y_pos :
    plt.text(a,x[a]+100,f"{x[a]}  ({round(x[a]/tot*100,2)}%)",ha='center',fontsize=14)
```
